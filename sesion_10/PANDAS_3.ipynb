```python
import pandas as pd

# Crea un diccionario
diccionario = {'nombre': ['Juan', 'Ana', 'Pedro', 'Marti', 'Joao', 'Miguel', 'David'],
               'edad': [25, 30, 35, 36, 24, 31, 21],
               'ciudad': ['Lima', 'Arequipa', 'Bogota', 'Lima', 'Lima', 'Caracas', 'Arequipa'],
               'pais': ['Peru', 'Peru', 'Colombia', 'Peru', 'Peru', 'Venezuela', 'Peru']}

# Crea un DataFrame a partir del diccionario
df = pd.DataFrame(diccionario)

# Muestra el DataFrame
print(df)
```

## ENFOCANDONOS EN ALGUNAS VARIABLES:
Muchas veces trabajaremos con bases de datos con muchas columnas o variables, pero casi siempre solo nos conviene centrarnos en la que nos sean de nuestra importancia, para ello podemos filtrar las columnas que vamos a trabajar:

```python
## En este caso vamos a trabajar con el nombre y la edad
df.loc[:,['nombre','pais']]
```

```python
## En este caso vamos a trabajar con el nombre y pais
df.loc[:,['nombre','ciudad']]
```

```python
df.iloc[:,[0,3]]
```

## FILAS:

```python
df.loc[ df['nombre']=='Miguel',:]
```

```python
df.loc[ df['edad']>32,:]
```

```python
df.iloc[ 0 ,:]
```

## CELDA

```python
df.loc[ df['nombre']=='Miguel', 'ciudad']
```

```python
df.loc[ df['pais']=='Peru', 'nombre']
```

## Cambiando el valor de las celdas

```python
## primero tenemos que acceder a las celdas y asignarle un nuevo valor:
df.loc[ df['pais']=='Peru', 'pais']='China'
```

```python
df.loc[ df['pais']=='China', 'pais']='Peru'
df
```
