## Uniedo Series con  pd.concat()

```python
#podemos unir series de forma vertical como horizontal usando este m√©todo
import pandas as pd
import numpy as np

years=[2010,2011,2012,2013,2014,2015]
PBI=[120,125,148,145,149]

#convirtiendolos en series:

years_serie=pd.Series(years)
PBI_serie=pd.Series(PBI)
```

```python
#Verticalmente
data=pd.concat([years_serie, PBI_serie] )
data
```

```python
#Horizontalmente
data=pd.concat([years_serie, PBI_serie], axis=1 )
data.rename(columns={0: "year", 1: "PBI"}, inplace=True) #creamos un Dataframe
data
```

### RECOMENDADA: De diccionarios a Dataframe

```python
data= { 'state': ['Ohio', 'Ohio', 'Ohio', 'Nevada', 'Nevada'],
'year': [2000,2001,2002,2001,2002],
'pop': [1.5, 1.7, 3.6, 2.4, 2.9]}
poblacion = pd.DataFrame(data) #Para crear un dataframe se usa pd.DataFrame()
poblacion
```
